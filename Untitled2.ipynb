```python
def advanced_clean(df):
    print("Before:", len(df))

    # حذف التكرار
    df = df.drop_duplicates()

    # حذف الفاضي
    df = df.dropna()

    # حذف المسافات الفاضية
    df["en"] = df["en"].str.strip()
    df["ar"] = df["ar"].str.strip()

    df = df[(df["en"] != "") & (df["ar"] != "")]

    print("After cleaning:", len(df))
    return df
```

```python
import string

def is_noisy(en, ar):
    # لو فيها أرقام كتير
    if sum(c.isdigit() for c in en) > 5:
        return True

    # لو فيها رموز كتير
    if sum(c in string.punctuation for c in en) > 10:
        return True

    # لو عربي فيه إنجليزي كتير (غلط ترجمة)
    if sum(c.isalpha() and c.isascii() for c in ar) > 5:
        return True

    return False


def remove_noise(df):
    clean_rows = []

    for _, row in df.iterrows():
        if not is_noisy(row["en"], row["ar"]):
            clean_rows.append(row)

    df = pd.DataFrame(clean_rows)
    print("After noise removal:", len(df))
    return df
```

```python
import re
import pandas as pd

def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"@\w+|#\w+", "", text)
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\d+", "", text)
    return text.strip()

def normalize_arabic(text):
    text = re.sub("[إأآا]", "ا", text)
    text = re.sub("ى", "ي", text)
    text = re.sub("ؤ", "و", text)
    text = re.sub("ئ", "ي", text)
    text = re.sub("ة", "ه", text)
    text = re.sub("[ًٌٍَُِّْـ]", "", text)
    return text

def preprocess_texts(ar_texts, en_texts, max_len=50, min_len=3):
    data = []
    for ar, en in zip(ar_texts, en_texts):
        ar_clean = normalize_arabic(clean_text(ar))
        en_clean = clean_text(en)
        if min_len <= len(ar_clean.split()) <= max_len and min_len <= len(en_clean.split()) <= max_len:
            data.append({"en": en_clean, "ar": ar_clean}) # Changed 'source' to 'en' and 'target' to 'ar'
    return pd.DataFrame(data)
```

```python
def load_data(ar_path, en_path, max_samples=70000):
    data = []
    with open(ar_path, encoding="utf-8") as ar, open(en_path, encoding="utf-8") as en:
        for i, (a, e) in enumerate(zip(ar, en)):
            if i >= max_samples:
                break
            data.append({
                "en": e.strip(),
                "ar": a.strip()
            })
    return data

data = load_data(ar_path, en_path)
```

```python
ar_lines = [item['ar'] for item in data]
en_lines = [item['en'] for item in data]

df = preprocess_texts(ar_lines, en_lines)

df = advanced_clean(df)
df = remove_noise(df)
```

```python
from transformers import MBart50TokenizerFast, MBartForConditionalGeneration

model_name = "facebook/mbart-large-50"

tokenizer = MBart50TokenizerFast.from_pretrained(model_name)
model = MBartForConditionalGeneration.from_pretrained(model_name)

tokenizer.src_lang = "ar_AR"
tokenizer.tgt_lang = "en_XX"
```

```python
data = load_data(ar_path, en_path)

from datasets import Dataset

dataset = Dataset.from_list([
    {"translation": {"ar": x["ar"], "en": x["en"]}}
    for x in data
])

dataset = dataset.train_test_split(test_size=0.1)

train_dataset = dataset["train"]
test_dataset = dataset["test"]

print("Done:", len(train_dataset), len(test_dataset))
```

```python
def preprocess(example):
    inputs = [x["ar"] for x in example["translation"]]
    targets = [x["en"] for x in example["translation"]]

    model_inputs = tokenizer(inputs, max_length=128, truncation=True, padding="max_length")

    labels = tokenizer(targets, max_length=128, truncation=True, padding="max_length")

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

train_dataset = train_dataset.map(preprocess, batched=True)
test_dataset = test_dataset.map(preprocess, batched=True)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=2, # Reduced batch size to mitigate OOM error
    per_device_eval_batch_size=4,
    learning_rate=3e-5,
    num_train_epochs=3,
    # evaluation_strategy="epoch", # Removed due to TypeError
    save_strategy="epoch",
    logging_dir="./logs",
    fp16=True
)
```

```python
bleu = evaluate.load("bleu")

def compute_metrics(eval_pred):
    preds, labels = eval_pred

    decoded_preds = tokenizer.batch_decode(preds, skip_special_tokens=True)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    decoded_labels = [[label] for label in decoded_labels]

    result = bleu.compute(predictions=decoded_preds, references=decoded_labels)
    return {"bleu": result["bleu"]}
```

```python
from transformers import Trainer, DataCollatorForSeq2Seq

data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer, model=model)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)
```

```python
trainer.train()
```

```python
def translate(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True)
    inputs = {k: v.to(model.device) for k, v in inputs.items()} # Move inputs to the model's device

    outputs = model.generate(
        **inputs,
        forced_bos_token_id=tokenizer.lang_code_to_id["en_XX"],
        num_beams=5,
        max_length=100
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)

print(translate("انا احب الذكاء الاصطناعي"))
```
